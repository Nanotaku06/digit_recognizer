```python
import numpy as np
import pandas as pd
from PIL import Image
```

```python
user_img = input("Type the image's address: ")
```

```python
file = Image.open(user_img)
numArray = np.asarray(file)
```

```python
#grayscale pngs give 2d arrays. Anything else gives 3d Arrays. Checking whether the img given resulted in a 2d or 3d array.
isgrayscale = False
try:
     newArr = numArray[4][5][0]
except IndexError:
    isgrayscale = True
    print("bruh")
```

```python
#if the img given was colored, then we grayscale it.
pngArr = []
if not isgrayscale:
   for i in numArray:
      row = []
      for j in i:
         rgb_sum = 0
         for k in j:
            rgb_sum += k
         row.append(rgb_sum/3) 
      pngArr.append(row) 
else: pngArr = numArray        
```

```python
#compressing the image into a 28px by 28px 2d array, so the AI can read it
ypx = 0
xpx = 0
newArr = []
moduloY = len(pngArr)%28
moduloX = len(pngArr[0])%28
for y in range(28):
    xpx = 0
    row = []
    for x in range(28):

        if(moduloY>0):
            isnotmultipleY = 1
            moduloY-=1
        else: isnotmultipleY = 0
        if(moduloX>0):
            isnotmultipleX = 1
            moduloX-=1
        else: isnotmultipleX = 0
        
        newPixel_sum = 0

        for i in range(ypx, int(len(pngArr)/28) + ypx + isnotmultipleY):

            for j in range(xpx, int(len(pngArr[0])/28) + xpx + isnotmultipleX ):
                newPixel_sum += pngArr[i][j]   
        row.append(int(newPixel_sum/((len(pngArr)/28) * (len(pngArr[0])/28))))
        
        xpx += int(len(pngArr[0])/28)
    ypx += int(len(pngArr)/28)
    newArr.append(row)
        
```

```python
#checking whether the background of the img is back or white.

bordermayority = 0

for i in range(0, len(newArr)):
    if i<2 or i> 25:
        for j in range(0, len(newArr[i])):
            if newArr[i][j] > 255/2:
                bordermayority+=1
            else:
                bordermayority -=1
    else:
        for j in range(0, 2):
            if newArr[i][j] > 255/2:
                bordermayority+=1
            else:
                bordermayority -=1
        for j in range(len(newArr[i])-2, len(newArr[i])):
            if newArr[i][j] > 255/2:
                bordermayority+=1
            else:
                bordermayority -=1        
                            
```

```python
d1Arr = []
temp = 1
for x in newArr: 
    for y in x:
        if(bordermayority>0):
            d1Arr.append(255-y)
        else:
            d1Arr.append(y)    
        temp = temp + 1
```
